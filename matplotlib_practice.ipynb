```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a Line Chart Using the Object-Oriented Interface

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# which is same as following

# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

# actually plot it.
ax.plot(x_axis, y_axis)
```

```python
# a quick way to graph something without refering a figure

# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Use plt.show()
# plt.show() should only be used once per session. If any display issue, "Restart and Clear Output" may fix it. 

# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Annotate Charts Using the MATLAB Method

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Annotate Charts Using the Object-Oriented Interface

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
ax.legend()
ax.set_xlabel("Date")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(5, 50)
ax.set_ylabel("Fare ($)")
ax.grid()
```

```python
# Assign 12 colors, one for each month.
#colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
colors = ["slateblue", "magenta", "lightblue", "green",
          "yellowgreen", "greenyellow", "yellow", "orange",
          "gold", "indianred", "tomato", "mistyrose"]

#explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
explode_values = list(map(lambda month: 0.2 if month == "July" else 0, x_axis))
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')

plt.show()
```

# Skill Drill

```python
# Skill Drill

# figure out highest percentage month in h1 (highest) and h2, (percentage, month) 
h1 = (0, "")
h2 = (0, "") 
for h in zip(y_axis, x_axis):
    if h[0] > h1[0]:
        t = h1
        h1 = h
        h = t
    if h[0] > h2[0]:
        h2 = h
explode_values = list(map(lambda m: 0.2 if (m == h1[1] or m == h2[1]) else 0, x_axis))
print(explode_values)
```

```python
# size 8x8
fig, ax = plt.subplots(figsize=(8, 8))

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green",
          "yellowgreen", "greenyellow", "yellow", "orange",
          "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis,
       labels=x_axis,
       shadow=True,
       explode=explode_values,
       colors=colors,
       startangle=90,
       counterclock=False,
       autopct="%.1f%%")

plt.show()
```
